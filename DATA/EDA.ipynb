```python
from scipy.stats import ttest_ind

# Assuming df is your DataFrame
def perform_t_test(feature, level1, level2):
    charges_level1 = df[df[feature] == level1]['charges']
    charges_level2 = df[df[feature] == level2]['charges']
    t_stat, p_value = ttest_ind(charges_level1, charges_level2)
    return t_stat, p_value

# Perform T-tests for various combinations
features = ['gender', 'smoker', 'region', 'medical_history', 'family_medical_history',
            'exercise_frequency', 'occupation', 'coverage_level']

for feature in features:
    unique_levels = df[feature].unique()
    for i in range(len(unique_levels)):
        for j in range(i+1, len(unique_levels)):
            level1 = unique_levels[i]
            level2 = unique_levels[j]
            t_stat, p_value = perform_t_test(feature, level1, level2)
            print(f"T-statistic ({level1} vs. {level2}): {t_stat}")
            print(f"P-value ({level1} vs. {level2}): {p_value}")
            if p_value < 0.05:
                print(f"The difference in charges between {level1} and {level2} is statistically significant.")
            else:
                print(f"There is no significant difference in charges between {level1} and {level2}.")
            print("\n")
```

```python
from scipy.stats import chi2_contingency

# Assuming df is your DataFrame
def perform_chi_square_test(feature1, feature2):
    contingency_table = pd.crosstab(df[feature1], df[feature2])
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    return chi2, p_value

# Perform Chi-square tests for various combinations
features = ['gender', 'smoker', 'region', 'medical_history', 'family_medical_history',
            'exercise_frequency', 'occupation', 'coverage_level']

for i in range(len(features)):
    for j in range(i+1, len(features)):
        feature1 = features[i]
        feature2 = features[j]
        chi2, p_value = perform_chi_square_test(feature1, feature2)
        print(f"Chi-square ({feature1} vs. {feature2}): {chi2}")
        print(f"P-value ({feature1} vs. {feature2}): {p_value}")
        if p_value < 0.05:
            print(f"There is a significant association between {feature1} and {feature2}.")
        else:
            print(f"There is no significant association between {feature1} and {feature2}.")
        print("\n")
```
