```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
#import xgboost as xg 
from sklearn.ensemble import AdaBoostRegressor
```

```python
def model_building(X_train,y_train):
    model_1=LinearRegression()
    model_1.fit(X_train,y_train)
    model_1.score(X_train,y_train)
    print(f'The LinearRegression score is {model_1.score(X_train,y_train)}')
    model_2=Lasso()
    model_2.fit(X_train,y_train)
    model_2.score(X_train,y_train)
    print(f'The Lasso score is {model_2.score(X_train,y_train)}')
    model_3=Ridge()
    model_3.fit(X_train,y_train)
    model_3.score(X_train,y_train)
    print(f'The Ridge score is {model_3.score(X_train,y_train)}')
    model_4=RandomForestRegressor()
    model_4.fit(X_train,y_train)
    model_4.score(X_train,y_train)
    print(f'The RandomForestRegressor score is {model_4.score(X_train,y_train)}')
    model_5=AdaBoostRegressor()
    model_5.fit(X_train,y_train)
    model_5.score(X_train,y_train)
    print(f'The AdaBoostRegressor score is {model_5.score(X_train,y_train)}')

    
```

```python
model_building(X_train,y_train)
```

```python
model_4=RandomForestRegressor()
model_4.fit(X_train,y_train)
model_4.score(X_train,y_train)
print(f'The RandomForestRegressor score is Training is {model_4.score(X_train,y_train)}')
model_4.score(X_test,y_test)
print(f'The RandomForestRegressor score is Training is  {model_4.score(X_test,y_test)}')
```

```python
param_grid = {
    'n_estimators': [50,150],
    'max_depth': [10, 20],
    'min_samples_split': [4,6],
    'min_samples_leaf': [2, 4],
    'max_features': ['sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_regressor = RandomForestRegressor()

# Create GridSearchCV with the specified parameter grid
grid_search = GridSearchCV(rf_regressor, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the model to find the best hyperparameters
grid_search.fit(X_train, y_train)
```

```python
grid_search.score(X_train, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Mean Score:", -grid_search.best_score_)
```

```python
grid_search.best_estimator_
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model.score(X_train,y_train)
```

```python
best_model.score(X_test,y_test)
```

```python
y_pred=best_model.predict(X_test)
```
